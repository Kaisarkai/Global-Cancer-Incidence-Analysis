#**Sorting**

## *Bubble sort*

```python
def bubbleSort(arr):
  n= len(arr)
  for i in range(n):
    for j in range(0, n-i-1):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]

if __name__ == "__main__":
  arr = cases
  bubbleSort(arr)
  print("Bubble Sorted of Cases in 2020")
  for i in range(len(arr)):
    print("%d" % arr[i], end =" ")
```

```python
def bulbblesort(arr):
  n = len(arr)
  for i in range(n):
    for j in range (0, n-i-1):
      if arr[j] > arr [j+1]:
        arr[j], arr[j+1] + arr[j+1], arr[j]

if __name__ == "__main__":
  arr = cancer
  bubbleSort(arr)
  print("Bubbble Sorted of Cancer in 2020")
  for i in range(len(arr)):
    print(arr[i], end = " ")
```

## *Insertion Sort*

```python
def insertion_sort(alist):
  for i in range (1, len(alist)):
    temp = alist[1]
    j = i - 1
    while (j >=0 and temp < alist[j]):
      alist[j + 1] = alist[j]
      j = j -1
    alist[j + 1] = temp

if __name__ == "__main__":
  alist = cases
  insertion_sort(alist)

  print("Insertion Sorted of Cases In 2020  :")
  for i in range(len(alist)):
    print("%d" % alist[i], end = " ")

```
